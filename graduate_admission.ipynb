### Converting TOEFL Scores to IELTS

Many universities provide approximate equivalence between TOEFL and IELTS scores, although there is **no official 1:1 conversion**.  

- **TOEFL total score:** 0-120  
- **IELTS total score:** 0-9 (in 0.5 increments)  

A commonly used approximation is:

\[
\text{IELTS} \approx \frac{\text{TOEFL Score}}{120} \times 9
\]

- Example conversions:  
  - TOEFL 120 → IELTS 9.0  
  - TOEFL 100 → IELTS 7.5  

This formula allows us to create an **approximate IELTS score** from TOEFL, which can be useful for **feature engineering** or further analysis in predicting graduate admissions.

```python
admission_df
```

```python
# Calculate IELTS
ielts_scores = (admission_df['TOEFL_Score'] / 120 * 9).round(1)

# Insert IELTS column at position 2 (after GRE_Score and TOEFL_Score)
admission_df.insert(2, 'IELTS', ielts_scores)

# Preview updated DataFrame
admission_df.head()
```

#### Distribution of GRE Score

### Understanding the Histogram

- The x-axis represents GRE score ranges.
- The y-axis represents the number of students (frequency) in each range.

The histogram shows the distribution of GRE scores among applicants.

#### Correlation Analysis

The correlation heatmap shows how strongly each feature is related to the others, particularly the Chance of Admit. 
This helps identify the most important factors influencing graduate admissions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = admission_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Graduate Admission Dataset')
plt.show()
```

```python
# How to Read the Heatmap
# The Scale (Legend): On the right, see a color bar.
# Dark Red (1.0): Perfect positive correlation (as one goes up, the other goes up perfectly).
# Light Orange/White (0.5–0.7): Moderate positive correlation.
# Blue (below 0.5): Weaker correlation.
# The Diagonal: Notice the line of 1.00 values running from the top-left to the bottom-right. This is because every variable is perfectly correlated with itself (e.g., GRE_Score vs. GRE_Score).
# The Intersections: To find the relationship between two factors, find where their row and column meet.
```

### Key Insights from the correlation matrix

The Best Predictors of Admission:
- CGPA (0.87): This is the highest value in the row. It means CGPA is the strongest predictor in your dataset.
- GRE_Score (0.80): This is the second strongest factor.
- TOEFL_Score & IELTS (0.79): These are also very high, showing language proficiency is a major priority.

The "Overlapping" Scores (Multicollinearity)
- TOEFL_Score and IELTS meet and can see 1.0. This suggests these two variables provide almost identical information. Usually, in a machine learning model, you might only need one of these because they are so perfectly correlated.
- Similarly, GRE and TOEFL (0.84) are strongly linked, likely because students who are good at standardized testing tend to score high on both.

Qualitative vs. Quantitative Factors
- SOP (0.68) and LOR (0.67) have a moderate relationship with admission. While important, they aren't as "mathematically" tied to the outcome as CGPA.
- Research (0.55) has the lowest correlation with the Chance of Admit compared to the test scores. This doesn't mean it isn't important, but it suggests that research alone won't get you in if your academic scores (CGPA/GRE) are low.

**Inference:**  
- CGPA is the strongest predictor of admission chance.
- GRE and TOEFL scores also positively influence admission probability.
- Research experience and strong SOP/LOR moderately improve admission chances.
- These correlations help identify which factors matter most for graduate admissions.
